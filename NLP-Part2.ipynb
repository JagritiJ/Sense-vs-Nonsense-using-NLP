```
#Lemmatisation
X_train['OptionA'] = X_train['OptionA'].apply(lambda x: " ".join([Word(word).lemmatize() for word in str(x).split()]))
X_train['OptionB'] = X_train['OptionB'].apply(lambda x: " ".join([Word(word).lemmatize() for word in str(x).split()]))
X_train['OptionC'] = X_train['OptionC'].apply(lambda x: " ".join([Word(word).lemmatize() for word in str(x).split()]))

X_test['OptionA'] = X_train['OptionA'].apply(lambda x: " ".join([Word(word).lemmatize() for word in str(x).split()]))
X_test['OptionB'] = X_train['OptionB'].apply(lambda x: " ".join([Word(word).lemmatize() for word in str(x).split()]))
X_test['OptionC'] = X_train['OptionC'].apply(lambda x: " ".join([Word(word).lemmatize() for word in str(x).split()]))
```

```
# Correcting Letter Repetitions
def de_repeat(text):
  pattern = re.compile(r"(.)\1{2,}")
  return pattern.sub(r"\1\1", text)

X_train['OptionA'] = X_train['OptionA'].apply(lambda x: " ".join(de_repeat(x) for x in str(x).split())) 
X_train['OptionB'] = X_train['OptionB'].apply(lambda x: " ".join(de_repeat(x) for x in str(x).split())) 
X_train['OptionC'] = X_train['OptionC'].apply(lambda x: " ".join(de_repeat(x) for x in str(x).split())) 

X_test['OptionA'] = X_train['OptionA'].apply(lambda x: " ".join(de_repeat(x) for x in str(x).split())) 
X_test['OptionB'] = X_train['OptionB'].apply(lambda x: " ".join(de_repeat(x) for x in str(x).split())) 
X_test['OptionC'] = X_train['OptionC'].apply(lambda x: " ".join(de_repeat(x) for x in str(x).split())) 
```

```
#This data contains numerical data in the sentences
mylist = []

for i in range(len(y_train)):
    # print(i)
    if(y_train.answer[i] == 'A'):
         mylist.append([X_train.OptionA[i], True])
         mylist.append([X_train.OptionB[i], False])
         mylist.append([X_train.OptionC[i], False])
    
    if(y_train.answer[i] == 'B'):
         mylist.append([X_train.OptionB[i], True])
         mylist.append([X_train.OptionA[i], False])
         mylist.append([X_train.OptionC[i], False])
                
    if(y_train.answer[i] == 'C'):
         mylist.append([X_train.OptionC[i], True])
         mylist.append([X_train.OptionA[i], False])
         mylist.append([X_train.OptionB[i], False])
                
  
print(mylist)

mylist2 = []

for i in range(len(y_test)):
    # print(i)
    if(y_test.answer[i] == 'A'):
         mylist2.append([X_test.OptionA[i], True])
         mylist2.append([X_test.OptionB[i], False])
         mylist2.append([X_test.OptionC[i], False])

    if(y_test.answer[i] == 'B'):
         mylist2.append([X_test.OptionB[i], True])
         mylist2.append([X_test.OptionA[i], False])
         mylist2.append([X_test.OptionC[i], False])
         
    if(y_test.answer[i] == 'C'):
         mylist2.append([X_test.OptionC[i], True])
         mylist2.append([X_test.OptionA[i], False])
         mylist2.append([X_test.OptionB[i], False])

print(mylist2)
```

```
myarr = np.array(mylist)
myarr1 = np.array(mylist2)
```

```
nltk.download('stopwords')
stop = stopwords.words('english')
```

```
tfidf = TfidfVectorizer(max_features=1000, stop_words=stop, lowercase=True, analyzer ='word', ngram_range=(1,3))
X_train_tfidf = tfidf.fit_transform(myarr[:,0])
X_test_tfidf = tfidf.fit_transform(myarr1[:,0])
X_train_tfidf
```

```
#Encoding output labels, mL can't handle words, convert to numbers
lbl_enc = preprocessing.LabelEncoder()
y_train = lbl_enc.fit_transform(myarr[:,1])
y_test = lbl_enc.fit_transform(myarr1[:,1])
```

```
from sklearn.naive_bayes import MultinomialNB
#Model: Multinomial Naive Bayes Classifier
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)
y_pred = nb.predict(X_test_tfidf)
print('naive bayes tfidf accuracy is %s' % accuracy_score(y_pred, y_test) )
```

```
#Using another Model - KNN 

from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train_tfidf, y_train) 
y_pred = neigh.predict(X_test_tfidf)
print('KNN tfidf accuracy is %s' % accuracy_score(y_pred, y_test) )
```
